```python
# Using Python, create fixed-width bins for the 'LotArea' feature in the dataset. Divide it into 5 equal intervals.
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/train_kaggle_housing.csv')
df['LotArea'].hist(bins=5)
```

```python
#Create custom bins for 'YearBuilt' based on historical periods: Pre-1900, 1900-1950, 1951-2000, Post-2000
bins_edges=[1872,1900,1950,2000,2010]
bins_labels=['Pre-1900', '1900-1950', '1951-2000', 'Post-2000']
df['YearBuilt'].plot.hist(bins=bins_edges)
```

```python
df['new_yearbuilt'] = pd.cut(df['YearBuilt'], bins=bins_edges, labels=bins_labels)
df['new_yearbuilt']
```
